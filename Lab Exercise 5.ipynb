The process below shows how to implement a simple linear regression analysis for a vehicle pricing scenario. Re-run it and take note of the R-sqaured value. Then. change the std value in np.random.normal to 1, 5 and 10. Observe the change in R-sqaured. How do you explain the change in R-sqaured value?

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 1, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

```python
vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 5, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

```python
vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 10, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

We increase the standard deviation from the random term that is added to the age of the car to determine its price, thus allowing randomness to have a bigger influence in the determination of the price and reducing the influence of the age of the car in the final price, which explains why we get a smaller value of R-squared (lower R-squared values indicate that a smaller proportion of the variance of the dependent variable can explained by the independent variable.)
