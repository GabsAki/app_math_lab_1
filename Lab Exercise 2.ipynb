```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# # of hours a AI student spends on their courses in a month
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()
```

```python
# Calculate and report the mean, std, and var of the data shown above.
print(f'Mean: {no_hours.mean()}')
print(f'Std: {no_hours.std()}')
print(f'Var: {no_hours.var()}')

# What is the theoretical relationship between std and var? Do the numbers reported in the previous step agree with your expectation?
print("The standard deviation is the squared root of the variance. The numbers reported do conform to this relationship")

# Begin changing the different parameters on the normal function, discuss your observations as you change each input parameter.
print("Changing the mean: Dislocates the curve (left if mean is lower, right if mean is higher).")
print("Changing the std: Changes the width the curve (a lower value widens the curve, while a higher value makes it thinner).")
print("Changing the number of samples: Changes the height of the curve (a lower value makes it taller, while a higher value makes it shorter).")
```
